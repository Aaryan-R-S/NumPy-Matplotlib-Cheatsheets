```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style 
```

PLOT

```python
x = [24,45,55]
y = [23,65,38]
x2 = [45,66,23]
y2 = [64,76,39]

style.use('ggplot')
plt.title("Sample Data")
plt.ylabel("This is y")
plt.xlabel("This is x")

plt.plot(x,y, color='y',label='first', linewidth=3)
plt.plot(x2,y2,color='r', label='second', linewidth=2)

plt.legend()   # for lable
plt.grid(True, color='k')    #  to make bg grids more opaque

plt.show()
```

BAR

```python
x = [24,45,55]
y = [23,65,38]
x2 = [45,66,23]
y2 = [64,76,39]

plt.xlabel('bar number')
plt.ylabel('bar height')

plt.bar(x,y, color='y',label='first')
plt.bar(x2,y2,color='r', label='second')

plt.show()
```

SCATTER

```python
x = [24,45,55]
y = [23,65,38]
x2 = [45,66,23]
y2 = [64,76,39]

plt.scatter(x,y, label='first', color='r')
plt.scatter(x2,y2, label='second', color='y')

plt.legend()   # for lable
plt.grid(True, color='k')    #  to make bg grids more opaque

plt.show()
```

HISTOGRAM

```python
x = [4,15,25,44,34,23,34]
y = [0,10,20,30,40,50]

plt.hist(x,y, histtype='bar')
plt.show()
```

PIE

```python
x = [2,4,8,6]
s =['LOSE','TIE','WON','DRAW']
cols = ['g','r','y','m']

plt.pie(x, labels=s, colors=cols, startangle=90, shadow=True, explode=(0,0.1,0,0), autopct='%1.1f%%')
plt.show()
```
